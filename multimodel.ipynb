```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets  import fetch_california_housing
cali_houses = fetch_california_housing()
```

```python
cali_houses.keys()
```

```python
cali_houses.feature_names
```

```python
cali_houses.data
```

```python
df = pd.DataFrame(cali_houses.data, columns=cali_houses.feature_names)
```

```python
df['Prices'] = cali_houses.target
```

```python
df
```

```python
X = df.drop('Prices', axis=1)
Y = df['Prices']
```

```python
for x in X.columns:
    plt.scatter(df[x].head(10000), Y.head(10000))
    plt.xlabel(x)
    plt.ylabel('Prices')
    plt.show()
```

```python
X.corr()
```

```python
X
```

```python
Y
```

```python
sns.heatmap(X.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
scaler = PolynomialFeatures(degree=2, include_bias=True)
X_Train = scaler.fit_transform(X_Train)
X_Test = scaler.transform(X_Test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_Train, Y_Train)
```

```python
Y_Pred = regressor.predict(X_Test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y_Pred,Y_Test)
```

```python
print(r2*100)
```

```python
residue = Y_Test - Y_Pred
residue
```

```python
sns.kdeplot(residue)
```

```python
plt.scatter(Y_Pred, residue)
```
